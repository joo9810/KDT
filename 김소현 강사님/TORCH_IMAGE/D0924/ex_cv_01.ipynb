### OpenCV 기본 사용법

```python
# 모듈 로딩
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# 이미지 관련 공통 변수
IMG_DIR = '../image/'
```

[1] 이미지 읽기 ==> ndarray

```python
IMG_FILE = IMG_DIR + 'logo.png'

im_data = cv2.imread(IMG_FILE)
gray_data = cv2.imread(IMG_FILE, flags=cv2.IMREAD_GRAYSCALE)
bread_cat1 = cv2.imread(IMG_DIR + 'bread_cat1.png')
bread_cat2 = cv2.imread(IMG_DIR + 'bread_cat2.jpg')
```

```python
# 이미지 데이터 속성
print('[im_data]')
print(f'shape: {im_data.shape}')
print(f'ndim: {im_data.ndim}D')

print('[gray_data]')
print(f'shape: {gray_data.shape}')
print(f'ndim: {gray_data.ndim}D')
```

[2] 이미지 화면 출력 => imshow()

```python
def image_show(image_data):
    # 이미지 창 띄우기
    cv2.imshow("[IM_DATA]", image_data)

    # 이미지 창 위에 이벤트 대기
    cv2.waitKey()

    while True:
        # 키보드 이벤트 대기하수
        key_code = cv2.waitKey()
        print(f'key_code => {key_code}, {chr(key_code)}')

        if chr(key_code) in ['x', 'X']:
            # 모든 창 닫기
            cv2.destroyAllWindows()
            # 맥북에서 이렇게 안하면 창이 안꺼진다는 이슈가 있음
            for i in range (1,5):
                cv2.waitKey(1)
            break
        elif chr(key_code) in ['s', 'S']:
            # 이미지 저장
            cv2.imwrite('copy.jpg', image_data)
            # 모든 창 닫기
            cv2.destroyAllWindows()
            # 맥북에서 이렇게 안하면 창이 안꺼진다는 이슈가 있음
            for i in range (1,5):
                cv2.waitKey(1)
            break

image_show(gray_data)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (15, 6))

bread_cat1_RGB = cv2.cvtColor(bread_cat1, cv2.COLOR_BGR2RGB)
bread_cat2_RGB = cv2.cvtColor(bread_cat2, cv2.COLOR_BGR2RGB)
axs[0].imshow(bread_cat1_RGB)
axs[1].imshow(bread_cat2_RGB)
plt.show()
```
