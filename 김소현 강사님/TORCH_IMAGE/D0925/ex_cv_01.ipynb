### 이미지 데이터 처리
- 채널 관련
    * 컬러 : 3개 채널 => (H, W, C) => (H, W, 3)
    * 흑백 : 1개 채널 => (H, W, C) => (H, W)

```python
# 모듈 로딩
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = '../image/'

bread_cat1 = cv2.imread(IMAGE_PATH+'bread_cat1.png')
bread_cat3 = cv2.imread(IMAGE_PATH+'bread_cat3.jpg')
```

```python
bread_cat1_RGB = cv2.cvtColor(bread_cat1, cv2.COLOR_BGR2RGB)
bread_cat3_RGB = cv2.cvtColor(bread_cat3, cv2.COLOR_BGR2RGB)

fig, axs = plt.subplots(1, 2, figsize = (10, 5))

axs[0].imshow(bread_cat1_RGB)
axs[1].imshow(bread_cat3_RGB)

plt.show()
```
